## 3) Printing the name of stocks using for loop

```python
# importing the csv library
import csv
  
# opening the csv file by specifying
# the location
# with the variable name as csv_file
with open('stock1.csv') as csv_file:
  
    # creating an object of csv reader
    # with the delimiter as ,
    csv_reader = csv.reader(csv_file, delimiter = ',')
  
    # list to store the names of columns
    list_of_column_names = []
  
    # loop to iterate thorugh the rows of csv
    for row in csv_reader:
  
        # adding the first row
        list_of_column_names.append(row)
  
        # breaking the loop after the
        # first iteration itself
        break
  
 # printing the result
print("List of Stock names : ",list_of_column_names[0])
```

## 9) Plotting RAW STOCK PRICES

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.plot(stock_df['Date'], stock_df['sp500'], color = 'purple')

ax.set(xlabel='Date',ylabel="Sp500",title="RAW STOCK PRICES")

plt.show()
```

```python
# Plot all the close prices

stock_df.plot(figsize=(10, 7))

# Show the legend
plt.legend()

# Define the label for the title of the figure
plt.title("Adjusted Close Price", fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('Price', fontsize=14)
plt.xlabel('Date', fontsize=14)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)

plt.show()
```
